```python
# Week 2 – Exploratory Data Analysis (EDA)
# Smart Healthcare Analytics System

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Load data
df = pd.read_csv("../data/diabetes.csv")   # or heart.csv
print("Shape:", df.shape)
df.head()

# 2️⃣ Basic info
df.info()
df.describe().T

# 3️⃣ Check for missing values
print("Missing values:\n", df.isna().sum())

# 4️⃣ Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()

# 5️⃣ Target distribution
target_col = "Outcome"   # for diabetes dataset; change if using heart.csv
sns.countplot(x=target_col, data=df)
plt.title(f"Target Variable Distribution: {target_col}")
plt.show()

# 6️⃣ Feature distributions
df.hist(figsize=(12,8), bins=20)
plt.suptitle("Feature Distributions", y=1.02)
plt.show()

# 7️⃣ Outlier visualization
numeric_cols = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numeric_cols])
plt.title("Boxplots of Numeric Features")
plt.xticks(rotation=45)
plt.show()

# 8️⃣ Pairplot (optional)
sns.pairplot(df, hue=target_col, diag_kind='hist', corner=True)
plt.show()

# 9️⃣ Summary stats save
summary = df.describe()
summary.to_csv("../docs/data_summary.csv", index=True)

print("EDA complete ✅")
```
